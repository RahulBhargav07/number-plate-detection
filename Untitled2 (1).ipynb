```python

FAQs – Agriculture AI & IoT App Project
--------------------------------------------------

1. What problem does this project solve?
Indian farmers face challenges due to lack of timely access to weather updates, mandi prices,
government schemes, and real-time crop health data. This project offers an integrated AI + IoT solution for precision farming.
--------------------------------------------------

2. Who are the intended users or customers?
The app is designed for farmers—small, marginal, and large—who need automated, AI-driven tools to boost productivity,
reduce risks, and receive timely support in their local language.
--------------------------------------------------

3. What is the value proposition of this app?
It unifies AI, IoT, and real-time data in a single mobile platform, providing farmers with personalized insights for irrigation,
pest detection, stress monitoring, mandi price updates, and government schemes—all in regional languages.
--------------------------------------------------

4. What does the product include?
A mobile app with 12 intelligent features linked to ESP32-based sensor kits and AI/ML models. It includes a voice-enabled multilingual
chatbot for ease of use, even by digitally inexperienced farmers.
--------------------------------------------------

5. What are the main objectives of the project?
- Automate irrigation using AI and sensors
- Detect plant diseases, stress, and pests
- Offer local alerts and advisory
- Predict seed quality and crop needs
- Assist farmers in availing government schemes
--------------------------------------------------

6. What risks and challenges are anticipated?
- Poor or no internet access in rural areas
- Sensor calibration and maintenance
- Varying levels of digital literacy among farmers
- Model generalization across crops and regions
--------------------------------------------------

7. What are the major tasks involved in building this system?
This is a multi-task AI system involving:
- Object detection (YOLOv11)
- Crop stress and pest classification (EfficientNet/XGBoost)
- Irrigation need regression
- Voice/text chatbot (Gemini & Whisper APIs)
--------------------------------------------------

8. How will success be measured?
- Model metrics: Accuracy, Precision, Recall, F1-Score
- Field metrics: Water savings, yield improvement
- User metrics: App usage, latency, farmer satisfaction
--------------------------------------------------

9. How was the model evaluated?
Using:
- Local and sample datasets
- Home-based sensor simulations
- Integration with the app interface
- Manual validation and farmer feedback
--------------------------------------------------

10. What kind of data is used?
- Weather: OpenWeather API
- Market: Mandi price APIs
- Sensors: Soil moisture, humidity, and gas sensors
- Images: Roboflow + custom dataset
- Text: Agricultural advisories (translated)
--------------------------------------------------

11. What is the implementation roadmap?
- Week 1–2: Set up ESP32 hardware kits and design app UI
- Week 3–4: Train ML models and link APIs
- Week 5–6: Integrate Firebase backend and sensor control
- Week 7–8: Conduct testing, collect feedback, fine-tune models
--------------------------------------------------

12. How will the system evolve post-deployment?
Models will retrain with new field data. Updates will be deployed via Azure and TrueFoundry.
Chatbot intelligence will grow using real user queries and Gemini’s LLM capabilities.
--------------------------------------------------

13. Who is building this system?
- 1 App Developer
- 1 Hardware Engineer
- 2 ML Engineers
--------------------------------------------------

14. What tools and platforms are used?
- Hosting: Azure, TrueFoundry
- Database: Firebase
- Vision Models: Roboflow + YOLOv11
- NLP/Voice: Gemini API, Whisper
- Hardware: ESP32 with soil, gas, and humidity sensors
--------------------------------------------------
```
